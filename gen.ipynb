```python
import pandas as pd
```

```python
import random
items= ['Bread','Milk','Diaper','Coke','Eggs','Rice','Biscuts','Potatoes','Onion','Orange','Apple','Musumbi','Watermelon']
data = {
        'TransactionId':[],
        'Items':[]
       }
for i in range(1,101):
    data['TransactionId'].append(i)
    purchases = []
    for i in range(2,random.randint(5,12)):
        x = random.choice(items)
        while x in data['Items']:
            x = random.choice(items)
        purchases.append(x)
    data['Items'].append(purchases)

df = pd.DataFrame(data)
    
data = []
for x in df.values:
    data.append([x[0]] + x[1])
df = pd.DataFrame(data)
df.fillna(value=pd.np.nan, inplace=True)
df.to_csv('sa.csv')    
```

```python
mat = {
        'id':[i for i in range(1,len(df)+1)]
}

for x in items:
    mat.update({x:[0 for i in range(1,len(df)+1)]})
```

```python
for i,x in enumerate(df.values):
    for y in x:
        if y in mat.keys():
            mat[y][i]=1
```
